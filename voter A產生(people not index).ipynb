```python
import pandas as pd
voterA=pd.read_csv("voterA.csv")
#Data=Data.set_index("People")
voterA
```

```python
import numpy as np
import random
```

```python

voterAArray=np.array(voterA)
voterAArray
```

```python
D_shape=np.array(voterAArray.shape)
num=D_shape[0]
dem=int(D_shape[1])
num
```

```python

i=1
while i<dem:
    random.seed()
    rad=np.array([random.uniform(-1.0,1.0)for g in range(num)])
    voterAArray[:,i]=rad
    i=i+1
voterAArray
```

```python
voterA=round(pd.DataFrame(voterAArray,columns=['People','A1 str','A2 str','B1 str','B1 str']),2)
voterA.to_csv('voterA.csv',index = 0)
voterA
```
